```python
import pandas as pd
from funcion import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def graficar(df, titulo, xlabel, ylabel):
    fig, axes = plt.subplots(figsize=(5,6))
    df.iloc[0,:].plot(ax=axes, color="red", lw=1, ls='-', marker='s',  markersize=4)
    plt.setp(axes.get_xticklabels(), rotation=45, ha='right')
    plt.grid(True)
    axes.set_title(titulo)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, rotation=90)
    plt.show()
```

```python
nombre = contar_archivo()[0]
datos = pd.ExcelFile(nombre)
```

```python
df = datos.parse()
df = df.iloc[:,2:13]
df.drop(columns = ['slope','C_2000_ast'], inplace=True)
```

```python
elevaciones = df.iloc[:,3:].copy()
```

```python
diferencias = pd.DataFrame()
diferencias['dif00'] = df[df.columns[3]]-df[df.columns[3]]
diferencias['dif00-05'] = df[df.columns[4]]-df[df.columns[3]]
diferencias['dif05-12'] = df[df.columns[5]]-df[df.columns[4]]
diferencias['dif12-14'] = df[df.columns[6]]-df[df.columns[5]]
diferencias['dif14-17'] = df[df.columns[7]]-df[df.columns[6]]
diferencias['dif17-18'] = df[df.columns[8]]-df[df.columns[7]]
```

```python
abatimientos = pd.DataFrame()
abatimientos['2000'] = df[df.columns[3]]-df[df.columns[3]]
abatimientos['2005'] = df[df.columns[4]]-df[df.columns[3]]
abatimientos['2012'] = df[df.columns[5]]-df[df.columns[3]]
abatimientos['2014'] = df[df.columns[6]]-df[df.columns[3]]
abatimientos['2017'] = df[df.columns[7]]-df[df.columns[3]]
abatimientos['2018'] = df[df.columns[8]]-df[df.columns[3]]
```

```python
#Elevaciones
graficar(elevaciones,'idpozo','año','elevacion')
```

```python
#abatimientos
graficar(abatimientos,'idpozo','año','elevacion')
```

```python
#Diferencias
graficar(diferencias,'idpozo','año','elevacion')
```
